## Problem Statement

You are opening a new Store at a particular location. Now, Given the Store Location, Area, Size and other params. Predict the overall revenue/Sale generation of the Store.

## Dataset Description

Dataset Details: The data has 8523 rows of 12 variables.

Columns:

- Item_Identifier: Unique product ID

- Item_Weight: Weight of product

- Item_Fat_Content: Whether the product is low fat or not

- Item_Visibility: The % of total display area of all products in a store allocated to the particular product

- Item_Type: The category to which the product belongs

- Item_MRP: Maximum Retail Price (list price) of the product

- Outlet_Identifier: Unique store ID

- Outlet_Establishment_Year: The year in which store was established

- Outlet_Size: The size of the store in terms of ground area covered

- Outlet_Location_Type: The type of city in which the store is located

- Outlet_Type: Whether the outlet is just a grocery store or some sort of supermarket

- Item_Outlet_Sales: Sales of the product in the particulat store. [This is the outcome variable to be predicted.]

## Null Value Treatment [Train]

Replacing Item_Weight null values with mean [12.857645184135976] because it is Numerical column.

```python
train["Item_Weight"].fillna(value = train["Item_Weight"].mean(), inplace = True)
```

Replacing Item_Weight null values with mode [Medium] because its is Categorical column.

```python
train["Outlet_Size"].fillna(value = train["Outlet_Size"].mode()[0], inplace = True)
```

```python
train.isnull().sum()
```

```python
print(train["Item_Fat_Content"].value_counts())
```

We Need to repalce 'LF' and 'low fat' with 'Low Fat' and 'reg' with 'Regular'.

```python
train = train.replace({'Item_Fat_Content' : { 'LF' : 'Low Fat', 'reg' : 'Regular', 'low fat' : 'Low Fat' }})
print(train["Item_Fat_Content"].value_counts())
```

```python
print(train["Item_Type"].value_counts())
```

### Insights (Item_Outlet_Sales Vs Item_Type)

- Fruits and Vegetables shows the highest average sales, indicating that it is a significant contributor to total sales.
- Snack Foods is another top performer with high average sales, suggesting a strong demand for snack items.
- Household also exhibits high average sales, indicating that household items are popular among customers.

### Strategic Insights:

- Focus on High Performers: To maximize sales, it would be beneficial to focus on top-performing categories like Fruits and Vegetables, Snack Foods, and Household.
- Improve Lower Performers: Investigating why certain categories like Breakfast, Hard Drinks, and Seafood have lower sales and developing strategies to boost their performance could be advantageous.
- Consistency: Ensuring consistent stock and availability of items in categories with low variation in sales can help maintain steady revenue streams.
- Manage Variability: For categories with high sales variation, analyzing the factors contributing to this variability (e.g., seasonal demand, supply issues) can help in developing strategies to stabilize sales.
